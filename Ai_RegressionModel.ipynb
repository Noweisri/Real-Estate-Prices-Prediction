```python
# Importing libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the datasets
dataset = pd.read_csv("aqardata.csv")
```

```python
print("Old shape  : " , dataset.shape)
dataset = dataset.dropna()
print("Dataset new shape after dropping null values :" , dataset.shape)
dataset.head(50)
```

```python
#converting clumns to numericals manualy
#dataset = dataset.drop('neighborhood', axis='columns')
dataset['mainlocation'].replace({'الرياض': 0, 'جدة': 1},inplace = True)
dataset['purpose'].replace({'سكني': 0, 'تجاري': 1 , 'سكني أو تجاري':2  , 'غير محدد':3},inplace = True)
dataset['sublocation'].replace({'وسط الرياض': 0, 'غرب الرياض': 1 , 'شمال الرياض':2 , 'شرق الرياض':3 , 'جنوب الرياض':4 , 'شمال جدة':5 , 'جنوب جدة':6},inplace = True)
dataset['frontage'].replace({'غرب': 0, 'شمال': 1 , 'شرق':2 , 'جنوب':3 , 'شمال غربي':4 , 'شمال شرقي':5 , 'جنوب غربي':6 , 'جنوب شرقي':7 , '4 شوارع':8 , '3 شوارع':9},inplace = True)
    

# using .get_dummies function to convert the categorical datatype to numerical and storing the returned dataFrame in a new variable df1
df1 = pd.get_dummies(dataset['neighborhood'])
# using pd.concat to concatenate the dataframes df and df1 and storing the concatenated dataFrame in df.
dataset = pd.concat([dataset , df1], axis=1).reindex(dataset.index)
# removing the column 'Purchased' from df as it is of no use now.
dataset.drop('neighborhood' , axis = 1 , inplace=True)

# Moving the target column (Pricepm) to the last
column_to_move = dataset.pop("Pricepm")
dataset.insert(142, "Pricepm", column_to_move)
    
# printing results
print("Dataset's new shape : " , dataset.shape)
dataset.head(40)
```

```python
#separating the dependent and independent values....

#X-Independent variable
X = dataset.iloc[ : , 0:141 ].values
print(X)

#y-dependent or target variable
y = dataset.iloc[ : , 142 ].values
print('\n',y)
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define the number of folds for cross-validation
kf = KFold(n_splits = 5 , shuffle = True , random_state = 93)

# Initialize arrays to store metrics for each fold
mae_scores = []
rmse_scores = []
r2_scores = []

# Perform K-fold cross-validation
for train_index, test_index in kf.split(X):
    # Split the dataset into training and testing sets for this fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Fit the model on the training set
    model.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = model.predict(X_test)
    
    # Calculate the MAE, RMSE, and R2 scores for this fold
    fold_mae = mean_absolute_error(y_test, y_pred)
    fold_rmse = mean_squared_error(y_test, y_pred, squared=False)
    fold_r2 = r2_score(y_test, y_pred)
    
    # Append the scores to the arrays
    mae_scores.append(fold_mae)
    rmse_scores.append(fold_rmse)
    r2_scores.append(fold_r2)

# Calculate the mean and standard deviation of the scores across all folds
mean_mae = np.mean(mae_scores)
std_mae = np.std(mae_scores)
mean_rmse = np.mean(rmse_scores)
std_rmse = np.std(rmse_scores)
mean_r2 = np.mean(r2_scores)
std_r2 = np.std(r2_scores)

# Print the results
print("Mean Absolute Error: {:.2f} +/- {:.2f}".format(mean_mae, std_mae))
print("Root Mean Squared Error: {:.2f} +/- {:.2f}".format(mean_rmse, std_rmse))
print("R-squared: {:.2f} +/- {:.2f}".format(mean_r2, std_r2))
print(r2_scores)
```
